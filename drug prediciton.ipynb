```python
le=LabelEncoder()
msex=le.fit_transform(mr['Sex'])
mbp=le.fit_transform(mr['BP'])
mchol=le.fit_transform(mr['Cholesterol'])
mna=le.fit_transform(mr['Na_to_K'])
mdrug=le.fit_transform(mr['Drug'])
mr.Sex=msex
mr.BP=mbp
mr.Cholesterol=mchol
mr.Na_to_K=mna
mr.Drug=mdrug
```

```python
mr
```

```python
mr.corr
```

```python
x=mr.drop(['Sex','Drug'],axis='columns')
print(type(x))
```

```python
x
```

```python
y=mr['Drug']
print(type(y))
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=60)
model.fit(x_train,y_train)
```

```python
model3=RandomForestClassifier(n_estimators=10)
model3.fit(x_train,y_train)
model3.score(x_test,y_test)
```

```python
model1=RandomForestClassifier(n_estimators=40)
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
model2=RandomForestClassifier(n_estimators=30)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```
